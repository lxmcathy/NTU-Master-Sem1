count frequency of words/stem/stem+stopwords

```python
import json
import random
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
import collections

recordcellword = ""
for x in reviewsCell:
    recordcellword = recordcellword+x
words = nltk.word_tokenize(recordcellword)
#print(words)
collections.Counter(words)
```

```python
# words
a = collections.Counter(words)
print(len(dict(a)))
a = a.most_common(20)
a = dict(a)
x=list(a.keys())
print(a)
y=list(a.values())
plt.bar(x,y)
print(y)
plt.xlabel('word')
plt.ylabel('number')
plt.title('word vs number')
plt.show()
```

```python
stoplistsymbol =[',','.','...','?','!','(',')','&',';','``','n\'t','\'s']
```
